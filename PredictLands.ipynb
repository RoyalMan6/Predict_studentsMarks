```python
import numpy as np
import pandas as pd
import numpy as np
import pandas as pd
dataset=pd.read_csv("xAPI-Edu-Data.csv")
```

```python
dataset.head()
```

```python
dataset.info()
```

```python
gradeM = {'L': 0, 'M': 1, 'H': 2}
dataset = dataset.replace({'Class': gradeM})
```

```python
dataset.head()
```

```python
dataset.info()
```

```python
dataset.nunique()
```

```python
dataset.drop(['Topic', 'GradeID','NationalITy'], axis=1, inplace= True)
```

```python
oh = pd.get_dummies(dataset,columns=["gender","PlaceofBirth","StageID","SectionID","Semester","Relation","ParentAnsweringSurvey","ParentschoolSatisfaction","StudentAbsenceDays"],prefix=["gender","PlaceofBirth","StageID","SectionID","Semester","Relation","ParentAnsweringSurvey","ParentschoolSatisfaction","StudentAbsenceDays"])
oh.columns
```

```python
oh = oh.reindex(columns = [col for col in oh.columns if col != 'Class'] + ['Class'])
oh
```

```python
oh.shape
```

```python
x = oh.iloc[:, :-1].values
y = oh.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=22)
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier()
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
print("ACCURACY OF THE MODEL: ", accuracy_score(y_test, y_pred))
print("F1 OF THE MODEL: ", f1_score(y_test,y_pred, average='macro'))
```
